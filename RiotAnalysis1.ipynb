```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data seeded from master tier player games
# so most if not all matches here are high elo
data = pd.read_pickle("D:/JSON/Data/ALL40000aggregated.pkl")
```

```python
sns.distplot(data['matchDuration'],kde=False,bins=60)
```

```python
# EUW, KR
# JP, NA

# Here we can see clearly that jp and kr have much higher surrender rates
# we can also see that na has longer lasting games

fig,axes = plt.subplots(nrows=2,ncols=2)
# remove the y ticks
for array in axes:
    for axis in array:
        axis.set_yticks([])
[[axis1,axis2],[axis3,axis4]] = axes
sns.distplot(data[data['region']=='EUW']['matchDuration'],ax=axis1, kde_kws={"label": "EUW"})
sns.distplot(data[data['region']=='KR']['matchDuration'],ax=axis2, kde_kws={"label": "KR"})
sns.distplot(data[data['region']=='JP']['matchDuration'],ax=axis3, kde_kws={"label": "JP"})
sns.distplot(data[data['region']=='NA']['matchDuration'],ax=axis4, kde_kws={"label": "NA"})
```

```python
# Not very clear, but shows the features on one plot
fig, ax = plt.subplots()
ax.set_yticks([])
sns.distplot(data[data['region']=='EUW']['matchDuration'],ax=ax, kde_kws={"label": "EUW"})
sns.distplot(data[data['region']=='KR']['matchDuration'],ax=ax, kde_kws={"label": "KR"})
sns.distplot(data[data['region']=='JP']['matchDuration'],ax=ax, kde_kws={"label": "JP"})
sns.distplot(data[data['region']=='NA']['matchDuration'],ax=ax, kde_kws={"label": "NA"})
```

```python
# get soloQ ranked games that take 20-23 minutes,
# these are (to a good approximation) surrendered games
surrendered = data[data['queueType']=='TEAM_BUILDER_RANKED_SOLO']
surrendered = surrendered.query('1200 <= matchDuration <= 1380')
```

```python
# Overall distribution of times in surrendered games
# Characteristic initial spike (quick surrenders)
# flattens off after a minute or so, but still significant
# this is seen for each region
sns.distplot(surrendered['matchDuration'])
```

```python
# Not very clear, but shows the features on one plot
fig, ax = plt.subplots()
ax.set_yticks([])
sns.distplot(surrendered[surrendered['region']=='EUW']['matchDuration'],ax=ax, kde_kws={"label": "EUW"})
sns.distplot(surrendered[surrendered['region']=='KR']['matchDuration'],ax=ax, kde_kws={"label": "KR"})
sns.distplot(surrendered[surrendered['region']=='JP']['matchDuration'],ax=ax, kde_kws={"label": "JP"})
sns.distplot(surrendered[surrendered['region']=='NA']['matchDuration'],ax=ax, kde_kws={"label": "NA"})
```

```python
# highest achieved ranking of all players to get an idea of ELO (in the surrendered games)
euw = surrendered[surrendered['region']=='EUW']
rankings = []
for match in euw['participants']:
    for person in match:
        rankings.append(person['highestAchievedSeasonTier'])
rankings = pd.DataFrame(rankings,columns=["EUW rank"])
plt.figure(figsize=(9, 6))
sns.countplot(x="EUW rank", data=rankings, palette="Greens_d",order=['CHALLENGER','MASTER','DIAMOND','PLATINUM','GOLD','SILVER','BRONZE','UNRANKED'])
```
